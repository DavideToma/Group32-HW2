# RQ 6 (bonus) : "Which day is the one with most passengers for each months?

```python
import pandas as pd
from datetime import datetime as dt
import matplotlib as plt
import seaborn as sns
```

```python
def extract_date(date):
    '''
        Take as input s string of a date and return the year and the hour
    '''
    parsed = dt.strptime(date,"%Y-%m-%d %H:%M:%S")
    parsed = dt.date(parsed)
    
    return parsed
```

```python
def weekday(date):
    date = date.weekday()
    return date
```

### January

```python
df = pd.read_csv("yellow_tripdata_2018-01.csv", sep = ",",usecols = ["tpep_pickup_datetime", "passenger_count"])
```

```python
df["weekday"] = df["tpep_pickup_datetime"].apply(extract_date)
```

```python
df["weekday"] = df["weekday"].apply(weekday)
```

```python
df = df.drop(['tpep_pickup_datetime'], axis = 1)
```

```python
k = sns.color_palette("husl", 7)
x_labels = ['Sunday','Monday','Tuesday', 'Wednesday', 'Thursday',  'Friday', 'Saturday']
ax = sns.countplot(data=df, x = 'weekday', palette = k)
sns.set(rc={'figure.figsize':(7,7)})
ax.set_xticklabels(x_labels)
```

The weekday where taxis are most used is wednesday.
